```
# importing libraries

import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error
```

```
# generate some random data using numpy which are evenly spaced from 0 to 2*pi
n = 300
x = np.linspace(0,2 * np.pi,n)

# creating tan values and cos values

cos_values = np.cos(x)
sin_values = np.sin(x)
```

```
# plotting the data
plt.figure(figsize=(20,10))
plt.plot(x,cos_values,color='b')
plt.plot(x,sin_values,linewidth=3)
```

```
# adding noise to data
noise= 0.5
new_sin_values = sin_values + np.random.uniform(-noise , noise , n)
new_cos_values = cos_values + np.random.uniform(-noise , noise , n)
```

```
# plotting data with noise
plt.figure(figsize=(20,10))
plt.plot(x,cos_values,color='b')
plt.plot(x,new_cos_values,color='r',linewidth=2)
plt.plot(x,sin_values,color='g')
plt.plot(x,new_sin_values,color='black',linewidth=2)
```

```
# calculating errors using both formulas
```

```
# 1. using numpy and mean_squared_error for sin

error = (1/n)*sum(np.power(sin_values-new_sin_values,2))
mse = mean_squared_error(sin_values,new_sin_values)
error , mse
```

```
# 1. using numpy and mean_squared_error for cos

error = (1/n)*sum(np.power(cos_values-new_cos_values,2))
mse = mean_squared_error(cos_values,new_cos_values)
error , mse
```
